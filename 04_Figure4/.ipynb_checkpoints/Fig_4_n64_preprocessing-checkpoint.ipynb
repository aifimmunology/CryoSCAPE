```r
so_gen <- function(x){
    pro <- strsplit(strsplit(x,'/per_sample_outs/')[[1]][2],'/count/')[[1]][1]


    exp <- strsplit(strsplit(x,'/outs/')[[1]][1],'/')[[1]][length(strsplit(strsplit(x,'/outs/')[[1]][1],'/')[[1]])]

    pro <- paste(exp,pro,sep='_')

    mtx <- Read10X_h5(x) # reading the matrix
    so <- CreateSeuratObject(mtx,project=pro) # creating the SO, using 'pro' 
    return(so) # adds the SO to the list for each sample's h5
    }
```

```r
# This lapply() function uses the list crated above to then read each h5 into memory as a Seurat Object (SO)
# the output is a list of seurat objects

fixed_so_w1 <- lapply(h5s_w1, so_gen)
fixed_so_w2 <- lapply(h5s_w2, so_gen)
fixed_so_w3 <- lapply(h5s_w3, so_gen)
fixed_so_w4 <- lapply(h5s_w4, so_gen)

head(fixed_so_w1[[1]][[]])
```

```r
fully_w1 <- Reduce(merge,fixed_so_w1) # this compresses your list of SOs into a single large SO
fully_w2 <- Reduce(merge,fixed_so_w2) # this compresses your list of SOs into a single large SO
fully_w3 <- Reduce(merge,fixed_so_w3) # this compresses your list of SOs into a single large SO
fully_w4 <- Reduce(merge,fixed_so_w4) # this compresses your list of SOs into a single large SO
```

```r
rm(fixed_so_w1)
rm(fixed_so_w2)
rm(fixed_so_w3)
rm(fixed_so_w4)
```

```r
fully_w1 <- JoinLayers(fully_w1)
fully_w2 <- JoinLayers(fully_w2)
fully_w3 <- JoinLayers(fully_w3)
fully_w4 <- JoinLayers(fully_w4)
```

```r
fully_w1[["percent.mt"]] <- PercentageFeatureSet(fully_w1, pattern = "^MT-")
fully_w2[["percent.mt"]] <- PercentageFeatureSet(fully_w2, pattern = "^MT-")
fully_w3[["percent.mt"]] <- PercentageFeatureSet(fully_w3, pattern = "^MT-")
fully_w4[["percent.mt"]] <- PercentageFeatureSet(fully_w4, pattern = "^MT-")
```

```r
options(repr.plot.width=22,repr.plot.height=5)

# Visualize QC metrics as a violin plot
p1 <- VlnPlot(fully_w1, features = c("percent.mt"), ncol = 1, pt.size=0)
p2 <- VlnPlot(fully_w2, features = c("percent.mt"), ncol = 1, pt.size=0)
p3 <- VlnPlot(fully_w3, features = c("percent.mt"), ncol = 1, pt.size=0)
p4 <- VlnPlot(fully_w4, features = c("percent.mt"), ncol = 1, pt.size=0)

p1 | p2
p3 | p3
```

### Clustering

```r
fully_w1 <- NormalizeData(fully_w1) %>% 
    FindVariableFeatures() %>% 
    ScaleData() %>% 
    RunPCA() %>% 
    RunUMAP(dims = 1:20) %>% 
    FindNeighbors(dims = 1:20) %>% 
    FindClusters(resolution = 0.5)
```

```r
fully_w2 <- NormalizeData(fully_w2) %>% 
    FindVariableFeatures() %>% 
    ScaleData() %>% 
    RunPCA() %>% 
    RunUMAP(dims = 1:20) %>% 
    FindNeighbors(dims = 1:20) %>% 
    FindClusters(resolution = 0.5)
```

```r
fully_w3 <- NormalizeData(fully_w3) %>% 
    FindVariableFeatures() %>% 
    ScaleData() %>% 
    RunPCA() %>% 
    RunUMAP(dims = 1:20) %>% 
    FindNeighbors(dims = 1:20) %>% 
    FindClusters(resolution = 0.5)
```

```r
fully_w4 <- NormalizeData(fully_w4) %>% 
    FindVariableFeatures() %>% 
    ScaleData() %>% 
    RunPCA() %>% 
    RunUMAP(dims = 1:20) %>% 
    FindNeighbors(dims = 1:20) %>% 
    FindClusters(resolution = 0.5)
```
