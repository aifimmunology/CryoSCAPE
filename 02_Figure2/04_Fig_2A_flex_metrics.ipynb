```python
df = pd.read_csv(di_ms['PB00618-03-02'])
df
```

```python
l = []
for j in di_ms.keys():
    # print(j)
    p = di_ms[j]
    df = pd.read_csv(p)
    # cells = int(df['Metric Value'][0].split(',')[0]+df['Metric Value'][0].split(',')[1]) # cells
    rpc = int(df['Metric Value'][4].split(',')[0]+df['Metric Value'][4].split(',')[1]) # Median Reads per Cell
    upc = int(df['Metric Value'][2].split(',')[0]+df['Metric Value'][2].split(',')[1]) # Median UMIs per Cell
    gpc = int(df['Metric Value'][3].split(',')[0]+df['Metric Value'][3].split(',')[1]) # Median Genes per Cell
    l.append([
        j,
        # cells,
        rpc,
        upc,
        gpc
    ])
l
```

```python
# df2 = pd.DataFrame(l, columns = ['Sample','Cells', 'Reads', 'UMIs', 'Genes'])
df2 = pd.DataFrame(l, columns = ['Sample1', 'reads', 'umis', 'genes'])
df2
```

```python
df2['sample'] = df2.Sample1.str.split('-', expand=True)[0]
df2
```

```python
df3 = df2[df2['sample'].isin(['PB00607', 'PB00609', 'PB00610'])]
df3
```
