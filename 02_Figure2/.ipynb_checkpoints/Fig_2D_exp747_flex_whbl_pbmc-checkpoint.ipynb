## 3prime vs FLEX, Beads vs FACS, WHBL vs PBMC

```r
library(dplyr)
library(Seurat)
library(patchwork)
library(H5weaver)
library(hise)
library(tidyverse)
library(SeuratObject)
```

### Reading in the H5s

```r
h5s_v3_1 <- list.files(
    path = '/home/jupyter/CS15_WHBL/Exp00747_Fixed_3prime/v3', 
    pattern = 'filtered_feature_bc_matrix.h5$', 
    full.names = TRUE, 
    recursive = TRUE
)

h5s_v3_2 <- list.files(
    path = '/home/jupyter/CS15_WHBL/Exp00747_Fixed_3prime/SubSample_v3_1', 
    pattern = 'filtered_feature_bc_matrix.h5$', 
    full.names = TRUE, 
    recursive = TRUE
)

h5s_v3_3 <- list.files(
    path = '/home/jupyter/CS15_WHBL/Exp00747_Fixed_3prime/SubSample_v3_2', 
    pattern = 'filtered_feature_bc_matrix.h5$', 
    full.names = TRUE, 
    recursive = TRUE
)
```

```r
h5s_fix <- list.files(
    path = '/home/jupyter/CS15_WHBL/Exp00747_Fixed_3prime/fixed', 
    pattern = 'filtered_feature_bc_matrix.h5$', 
    full.names = TRUE, 
    recursive = TRUE
)

length(h5s_fix)
h5s_fix
```

```r
h5s_v3_1
```

```r
v3_so_1 <- lapply(h5s_v3_1, function(h5){
    x <- h5[[1]]
    x <- strsplit(x,'/outs/')[[1]][1]
    l <- length(strsplit(x ,'/')[[1]])
    x <- strsplit(x ,'/')[[1]][l]
    if(strsplit(x,'_')[[1]][4] == 'neg'){
        mtx <- Read10X_h5(h5)
        so <- CreateSeuratObject(mtx,project=x)
        return(so)}
    })
```

```r
v3_so_2 <- lapply(h5s_v3_2, function(h5){
    x <- h5[[1]]
    x <- strsplit(x,'/count/')[[1]][1]
    l <- length(strsplit(x ,'/')[[1]])
    x <- strsplit(x ,'/')[[1]][l]

    mtx <- Read10X_h5(h5)
    so <- CreateSeuratObject(mtx,project=x)
    return(so)})

v3_so_3 <- lapply(h5s_v3_3, function(h5){
    x <- h5[[1]]
    x <- strsplit(x,'/count/')[[1]][1]
    l <- length(strsplit(x ,'/')[[1]])
    x <- strsplit(x ,'/')[[1]][l]

    mtx <- Read10X_h5(h5)
    so <- CreateSeuratObject(mtx,project=x)
    return(so)})
```

```r
fixed_so <- lapply(h5s_fix, function(h5){
    x <- h5[[1]]
    x <- strsplit(x,'/count/')[[1]][1]
    l <- length(strsplit(x ,'/')[[1]])
    x <- strsplit(x ,'/')[[1]][l]
    if(strsplit(x,'_')[[1]][4] == 'neg'){
        mtx <- Read10X_h5(h5)
        so <- CreateSeuratObject(mtx,project=x)
        return(so)}
    })

fixed_so
```

```r
v3_so_1 <- Reduce(merge, v3_so_1)
v3_so_2 <- Reduce(merge, v3_so_2)
v3_so_3 <- Reduce(merge, v3_so_3)
fixed_so <- Reduce(merge, fixed_so)

v3_so_1 <- JoinLayers(v3_so_1)
v3_so_2 <- JoinLayers(v3_so_2)
v3_so_3 <- JoinLayers(v3_so_3)
fixed_so <- JoinLayers(fixed_so)
```

```r
v3_so_1$run <- 'v3_sub0'
v3_so_2$run <- 'v3_sub1'
v3_so_3$run <- 'v3_sub2'
fixed_so$run <- 'fix'
```

```r
options(repr.plot.width=15,repr.plot.height=8) 

pal2 <- c(sage, yl_or)

fixed_so$sample2 <- factor(fixed_so$sample2, levels = c('Leukopak','WHBL-1','WHBL-2','WHBL-3'))

Idents(fixed_so) <- 'sample2'

p1 <- subset(fixed_so, subset=nCount_RNA<50000) %>% 
    VlnPlot(features = "nFeature_RNA", ncol = 1,pt.size=0, split.by='clean', split.plot=TRUE) + #, cols=pal2) +
    ggtitle('Leukopak vs WHBL / FACS vs Bead Clean') +
    theme(
        plot.title = element_text(color="black", size=28, face='bold'),
        legend.key.size = unit(1.5, 'cm'),
        axis.title.y = element_text(size=22, face='bold'),
        legend.text = element_text(face='bold',size=16),
        legend.title = element_text(face='bold',size=16),
        axis.text.x = element_text(face='bold',angle = 45, size=16, vjust = 1, hjust = 1),
        axis.text.y = element_text(face='bold',size=16),
    ) + stat_summary(fun=median, geom="point", position = position_dodge(0.7),size=3, color="black") + 
    labs(
        x=NULL,
        y="Genes per Cell",
        color = '10X Chemistry'        
    )

d1 <- ggplot_build(p1)$data[[2]]

p1 <- p1 + annotate(
    "text",
    x = d1$x,  # Adjust the x-coordinate based on your groups
    y = tapply(d1$y,d1$group,mean)+1,  # Adjust the y-coordinate for text positioning
    label = round(tapply(d1$y,d1$group,mean),0),  # Display rounded median values
    vjust = 1.5,
    hjust = 0.5,
    size = 6,
    color = "black"
    )
p1 + scale_fill_manual(name='Sample\nPurification',values=pal2)
```
