## Creating Seurat Objects from h5 outs from Cellranger

### Reading h5 files into memory

```r
# This Little Script walks thru your directories starting in that associated with 'path'
# adds the path to a list for future reading into Seurat Objects

h5s <- list.files(
    # Place your analysis in a folder by number experiment, or whatever might work for you
    # It helps if this worksheet in in the same folder
    path = '/home/jupyter/CS15_WHBL/CWB_Paper/01_Final_Data/02_Data/2E_exp823/Exp00823_w2', # fill out the path variable here to point toward whatever directory you have your h5 files in
    pattern = 'filtered_feature_bc_matrix.h5$', # change this to 'sample_raw_feature_bc_matrix.h5$' if you need the raw h5s
    full.names = TRUE, 
    recursive = TRUE
)

length(h5s)
h5s
```

### Creating Seurat Objects

```r
# This lapply() function uses the list crated above to then read each h5 into memory as a Seurat Object (SO)
# the output is a list of seurat objects

fixed_so <- lapply(h5s, function(x){
    pro <- strsplit(strsplit(x,'/per_sample_outs/')[[1]][2],'/count/')[[1]][1]
    # pro is short for 'project', used in the CreateSeuratObject() function to label each SO something helpful. 
    # As long as your h5 path has something helpful (and unique!) to name it 
    exp <- strsplit(strsplit(x,'/outs/')[[1]][1],'/')[[1]][length(strsplit(strsplit(x,'/outs/')[[1]][1],'/')[[1]])]
    # Exp refers to the name of the Cellranger runs, ie the 10x GEM well, since each well must be ran on it's own cellranger call
    pro <- paste(exp,pro,sep='_')
    # Concatenting 'exp' and 'pro' so each well is distinct as is each sample
    mtx <- Read10X_h5(x) # reading the matrix
    so <- CreateSeuratObject(mtx,project=pro) # creating the SO, using 'pro' 
    return(so) # adds the SO to the list for each sample's h5
    })

head(fixed_so[[1]][[]])
```

```r
fully <- Reduce(merge,fixed_so) # this compresses your list of SOs into a single large SO
```

```r
head(fully[[]])
```

```r
# The [[ operator can add columns to object metadata. This is a great place to stash QC stats
fully[["percent.mt"]] <- PercentageFeatureSet(fully, pattern = "^MT-")
```

```r
options(repr.plot.width=12,repr.plot.height=6) 

Idents(fully)<-'orig.ident' # set the metadata column to be used in plotting, use whatever one makes sense

VlnPlot(subset(fully, subset=nCount_RNA<10000), features = "percent.mt", ncol = 1,pt.size=0)
```

```r
fully <- subset(fully, subset = percent.mt < 5)
```

### Initial QC: UMIs and Genes per Cell

```r
# changing these parameters changes the size of the plot, its a good idea to include them in every cell that produces a plot
options(repr.plot.width=18,repr.plot.height=8) 

Idents(fully)<-'orig.ident' # set the metadata column to be used in plotting, use whatever one makes sense

VlnPlot(subset(fully, subset=nCount_RNA<10000), features = "nCount_RNA", ncol = 1,pt.size=0) +
    geom_hline(yintercept=median(fully$nCount_RNA)) +
    ggtitle('UMIs per Cell')
```

```r
options(repr.plot.width=18,repr.plot.height=8)

Idents(fully)<-'orig.ident'

VlnPlot(fully, features = "nFeature_RNA", ncol = 1,pt.size=0) +
    geom_hline(yintercept=median(fully$nFeature_RNA)) +
    ggtitle('Genes per Cell')
```

### Normalizing, running PCA and UMAP clustering

```r
fully <- NormalizeData(fully) %>% 
    FindVariableFeatures() %>% 
    ScaleData() %>% 
    RunPCA() %>% 
    RunUMAP(dims = 1:20) %>% 
    FindNeighbors(dims = 1:20) %>% 
    FindClusters(resolution = 0.5)
```
