## Creating Seurat Objects from h5 outs from Cellranger

### Reading h5 files into memory

```r
# This Little Script walks thru your directories starting in that associated with 'path'
# adds the path to a list for future reading into Seurat Objects

h5s <- list.files(
    # Place your analysis in a folder by number experiment, or whatever might work for you
    # It helps if this worksheet in in the same folder
    path = '/home/jupyter/CS15_WHBL/CWB_Paper/01_Final_Data/03_Data/exp888_scrublet', # fill out the path variable here to point toward whatever directory you have your h5 files in
    pattern = 'filtered_feature_bc_matrix.h5$', # change this to 'sample_raw_feature_bc_matrix.h5$' if you need the raw h5s
    full.names = TRUE, 
    recursive = TRUE
)

length(h5s)
h5s
```

### Creating Seurat Objects

```r
# # This lapply() function uses the list crated above to then read each h5 into memory as a Seurat Object (SO)
# # the output is a list of seurat objects

fixed_so <- lapply(h5s, function(x){
    pro <- strsplit(strsplit(x,'/exp888_scrublet/')[[1]][2],'_sample_')[[1]][1]

    # exp <- strsplit(strsplit(x,'/outs/')[[1]][1],'/')[[1]][length(strsplit(strsplit(x,'/outs/')[[1]][1],'/')[[1]])]

    # pro <- paste(exp,pro,sep='_')

    mtx <- Read10X_h5(x) # reading the matrix
    so <- CreateSeuratObject(mtx,project=pro) # creating the SO, using 'pro' 
    return(so) # adds the SO to the list for each sample's h5
    })

head(fixed_so[[1]][[]])
```

```r
# This Little Script walks thru your directories starting in that associated with 'path'
# adds the path to a list for future reading into Seurat Objects

scrub <- list.files(
    # Place your analysis in a folder by number experiment, or whatever might work for you
    # It helps if this worksheet in in the same folder
    path = '/home/jupyter/CS15_WHBL/CWB_Paper/01_Final_Data/03_Data/exp888_scrublet', 
    pattern = '_2.csv$', # change this to 'sample_raw_feature_bc_matrix.h5$' if you need the raw h5s
    full.names = TRUE, 
    recursive = TRUE
)
scrub
```

```r
scrubs <- lapply(scrub,read.csv)
```

```r
head(scrubs[[1]])
```

```r
scrubs <- scrubs %>% lapply(function(x){
    rownames(x) <- x$X
    x$X <- NULL
    x$Barcodes <- rownames(x)
    return(x)
})
```

```r
head(scrubs[[1]])
```

```r
head(fixed_so[[1]])
```

```r
for (i in c(1:16)){
    # so <- fixed_so_w1[[i]]
    # scrub <- scrubs[[i]]
    # fixed_so[[i]]$Doublet_Score <- scrubs[[i]]$Doublet_Score
    # fixed_so[[i]]$Predicted_Doublet <- scrubs[[i]]$Predicted_Doublet
    
    fixed_so[[i]][[]]$Barcodes <- rownames(scrubs[[i]])
    
    fixed_so[[i]][[]] <- left_join(fixed_so[[i]][[]], scrubs[[i]], by = 'Barcodes') # , by='row.names', all=TRUE
}
```

```r
head(fixed_so[[1]])
```

```r
fully <- Reduce(merge,fixed_so) # this compresses your list of SOs into a single large SO
rm(fixed_so)
fully <-  JoinLayers(fully)
```

### Normalizing, running PCA and UMAP clustering

```r
fully <- NormalizeData(fully) %>% 
    FindVariableFeatures() %>% 
    ScaleData() %>% 
    RunPCA() %>% 
    RunUMAP(dims = 1:20) %>% 
    FindNeighbors(dims = 1:20) %>% 
    FindClusters(resolution = 0.5)
```

### Saving the SO to a file that can be read into memory later.


```r
ref <- LoadH5Seurat(file = '/home/jupyter/pbmc_multimodal.h5seurat')
```

```r
anchors <- FindTransferAnchors(
    reference = ref,
    query = fully,
    normalization.method = "SCT",
    reference.reduction = "spca",
    dims = 1:50
)
```

```r
fully_cl <- TransferData(
    anchorset = anchors, 
    reference = ref, 
    query = fully,
    refdata = list(
        celltype.l1 = "celltype.l1",
        celltype.l2 = "celltype.l2",
        celltype.l3 = "celltype.l3",
        predicted_ADT = 'ADT'
    )
)
```

```r
fully <- fully_cl
rm(fully_cl)
```
