## Helper functions (See https://github.com/AllenInstitute/aifi-swanson-teaseq)

```r
# Finds the overlapping TSS from the input ref file and the input sample files and returns a GRanges object
filter_gr <- function(query_gr,
                      target_gr,
                      mode = "remove",
                      ignore_strand = TRUE) {
  
  overlapping_fragments <- unique(
    S4Vectors::queryHits(
      GenomicRanges::findOverlaps(query_gr,
                                  target_gr,
                                  ignore.strand = ignore_strand)))
  
  if(mode == "remove") {
    query_gr[-overlapping_fragments]
  } else if(mode == "keep") {
    query_gr[overlapping_fragments]
  }
}
```

```r
# This function helps to pileup the TSS fragments based on the target window of 2000bp downstream and upstream of the TSS
pileup_fragments <- function(query_gr,
                            target_gr,
                            target_width = NULL) {
  
  if(is.null(target_width)) {
    target_width <- width(target_gr)[1]
  }
  target_gr <- GenomicRanges::resize(target_gr, 
                                     width = target_width,
                                     fix = "center")
  
  pos_target <- target_gr[strand(target_gr) == "+"]
  neg_target <- target_gr[strand(target_gr) == "-"]
  
  pos_ol <- findOverlaps(query_gr, pos_target, ignore.strand = TRUE)
  neg_ol <- findOverlaps(query_gr, neg_target, ignore.strand = TRUE)
  
  pos_rel <- data.frame(start = start(query_gr)[S4Vectors::queryHits(pos_ol)] - start(pos_target)[S4Vectors::subjectHits(pos_ol)],
                        end = end(query_gr)[S4Vectors::queryHits(pos_ol)] - start(pos_target)[S4Vectors::subjectHits(pos_ol)])
  pos_rel$start[pos_rel$start < 0] <- 0
  pos_rel$end[pos_rel$end > target_width] <- target_width
  
  pos_cov <- integer(target_width + 1)
  for(i in 1:nrow(pos_rel)) {
    pos_cov[(pos_rel$start[i] + 1):(pos_rel$end[i] + 1)] <- pos_cov[(pos_rel$start[i] + 1):(pos_rel$end[i] + 1)] + 1
  }
  
  neg_rel <- data.frame(start = end(neg_target)[S4Vectors::subjectHits(neg_ol)] - end(query_gr)[S4Vectors::queryHits(neg_ol)],
                        end = end(neg_target)[S4Vectors::subjectHits(neg_ol)] - start(query_gr)[S4Vectors::queryHits(neg_ol)])
  neg_rel$start[neg_rel$start < 0] <- 0
  neg_rel$end[neg_rel$end > target_width] <- target_width
  
  neg_cov <- integer(target_width + 1)
  for(i in 1:nrow(neg_rel)) {
    neg_cov[(neg_rel$start[i] + 1):(neg_rel$end[i] + 1)] <- neg_cov[(neg_rel$start[i] + 1):(neg_rel$end[i] + 1)] + 1
  }
  
  data.frame(pos = 0:(target_width),
             cov = pos_cov + neg_cov)
  
}
```

## Examples from TEA-seq

- Read fragments.tsv files and load them into GRanges
- Filters to keep only fragments that overlap TSS

```r
# Path to input directory
input_dir <- '/home/jupyter/wb_exp-940-967/cs15_ficoll/cs15_ficoll_bed_files/14M_bed_files'

# List downsampled bed files in input directory
bed_files <- list.files(input_dir, full.names = TRUE)

# Show file list
print(bed_files)
```

```r
# Get the TSS fragment list (calculate coverage at each position)
if(!file.exists("Figure_1_tss_fragment_list.rds")){
  tss_fragment_list <- map(
      bed_files,
    function(bed_file) {
     print(paste("Reading and filtering",bed_file))
     
     frags_df <- read_tsv(bed_file, 
                          col_names = c("chr","start","end","score","strand"))
       # If bed chr, start, end, name, score, strand
     frags_gr <- GRanges(seqnames = frags_df$chr,
                         IRanges(frags_df$start, frags_df$end))
     frags_gr <- unique(frags_gr)
     
     tss_frags_gr <- filter_gr(frags_gr,
                               tss_2kb,
                               mode = "keep")
     tss_frags_gr
    })
  saveRDS(tss_fragment_list,
          "Figure_1_tss_fragment_list.rds")
} else {
  tss_fragment_list <- readRDS("Figure_1_tss_fragment_list.rds")
}
```

```r
head(tss_fragment_list)
```

- Compute how many fragments overlap each position in the target_gr object
- Return a data.frame with position and coverage

```r
# Get a dataframe of positon and coverage for each bed file
if(!file.exists("Figure_1_tss_pileups_list.rds")) {
  tss_pileups <- map(
      tss_fragment_list,
      function(tss_fragments) {
        pileup_fragments(query_gr = tss_fragments,
                        target_gr = tss_2kb,
                        target_width = 4000)
      })
  saveRDS(tss_pileups,
          "Figure_1_tss_pileups_list.rds")
} else {
  tss_pileups <- readRDS("Figure_1_tss_pileups_list.rds")
}
```

### Prepare results for plots

```r
# Assign color and DonorID_Condition to each dataframe in order that the files were read in
color_mapping <- c(
  "#619CFF",
  "#F8766D",
  "#619CFF",
  "#F8766D",
  "#619CFF",
  "#F8766D"
)

donor_mapping <- c('Donor1_CS15', 'Donor1_Ficoll', 'Donor3_CS15', 'Donor3_Ficoll', 'Donor2_CS15', 'Donor2_Ficoll')

for(i in 1:length(tss_pileups)) {
  tss_pileups[[i]]$well_color <- color_mapping[i]
  tss_pileups[[i]]$well_label <- donor_mapping[i]
}

tss_pileups
```

```r
# Combine dataframes for plotting
tss_pileup_df <- list_rbind(tss_pileups)
tss_pileup_df
```

```r
# Manual normalization of total number of fragments
normalize_coverage <- function(tss_pileup_df) {
  # Calculate the total coverage
  total_coverage <- sum(tss_pileup_df$cov)
  
  # Normalize the coverage by dividing each coverage value by the total coverage
  normalized_coverage <- tss_pileup_df$cov / total_coverage
  
  return(normalized_coverage)
}

# Apply normalization
normalized_data <- normalize_coverage(tss_pileup_df)

# Add normalized coverage to the combined df
tss_pileup_df$normalized_coverage <- normalized_data

# Check that the normalized values were added to the df
print(tss_pileup_df)
```

```r
# Plot the normalized TSS df as a pileup plot of coverage with relative position 2kbp downstream and upstream of the TSS
normalized_tss_fp_pileup_plot <- ggplot(data = tss_pileup_df) +
  geom_line(aes(x = pos, y = normalized_coverage,
                group = well_label,
                color = well_color,
                linetype = well_label),
            size = 0.5) +
  scale_color_identity() +
  scale_linetype_manual(values = c("solid", "solid", "dotted", "dotted","dashed", "dashed"), # Define line types by donor in order of file list
                        breaks = unique(tss_pileup_df$well_label)) + 
  ylab("Mean footprints per cell") +
  scale_x_continuous("Pos. Relative to TSS",
                     breaks = seq(0, 4000, 1000),
                     labels = seq(-2000, 2000, 1000)) +
  theme_bw(base_size = 7) +
  theme(panel.grid.major = element_line(colour = "gray80",
                                        size = 0.2),
        panel.grid.minor = element_blank(),
        panel.border = element_blank(),
        axis.ticks = element_blank(),
        axis.text = element_text(colour = "black")) +
  labs(linetype = "Well Label")

# Print the plot and legend
normalized_tss_fp_pileup_plot + theme(legend.position = "right")
```
