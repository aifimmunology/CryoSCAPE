```r
deg <- function(x){
    x$genes <- rownames(x)
    x <- x[x$p_val_adj<0.05,]
    return(x)
}
```

```r
add.celltypes <- function(marker_list){
    new_marker_list <- names(marker_list) %>% lapply(function(name){
        marker_list[[name]]$celltype <- name
        return(marker_list[[name]])
    })
    return(new_marker_list)
}
```

```r
HeatMap <- function(object, l1.celltype, Stim){
    avg_exp <- object %>% 
        subset(subset = predicted.celltype.l1 == l1.celltype) %>% 
        subset(subset = stim %in% c(Stim, 'No_Stim')) %>% 
        AverageExpression(group.by = 'donor_stim', assay = 'RNA', layer = 'data')

    avg_exp <- as.data.frame(avg_exp$RNA)
    avg_exp <- avg_exp[rowSums(avg_exp)>0,]
    
    return(as.matrix(avg_exp))
}
```
